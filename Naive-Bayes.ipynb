```python
# Importing libraries 
import pandas as panda
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Data Import
rawData = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
attribute = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Class']
dataFrame = panda.read_csv(rawData, names=attribute)
```

```python
#Step 1
x = dataFrame.iloc[:, :-1] #Features
y = dataFrame.iloc[:, 4] #Class
```

```python
#Step 2
scaler = StandardScaler()
train = scaler.fit_transform(x)
```

```python
#Step 3
X_train, X_test, y_train,y_test = train_test_split(train,y,test_size=0.30)
```

```python
#Step 4
gnb = DecisionTreeClassifier()
gnb.fit(X_train, y_train)
gnb = KNeighborsClassifier(n_neighbors=7)
gnb.fit(X_train, y_train)
```

```python
#Step 5
pred = gnb.predict(X_test)
gnb.score(X_test,y_test)
```
