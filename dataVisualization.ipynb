```python
# Importing libraries 
import pandas as panda
import matplotlib.pyplot as matplot

# Data Import
rawData = "dataset/iris.csv"
dataFrame = panda.read_csv(rawData, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])

dataFrame.head(1)
```

```python
# Create a figure and axis 
fig, ax = matplot.subplots()

# Scatter the sepal_length against the sepal_width
ax.scatter(dataFrame['sepal_length'], dataFrame['sepal_width'])

# Set a title and labels

ax.set_title('Scatter Diagram')
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
```

```python
# Create color dictionary
colors = {'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}
# Create a figure and axis
fig, ax = matplot.subplots()

for i in range(len(dataFrame['sepal_length'])):
    ax.scatter(dataFrame['sepal_length'][i], dataFrame['sepal_width'][i],color=colors[dataFrame['class'][i]])

# Set a title and labels
ax.set_title('Scatter diagram | Colour')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')

matplot.show()
```

```python
# Get columns to plot
columns = dataFrame.columns.drop(['class'])
# Create x data
x_data = range(0, dataFrame.shape[0])
# Create figure and axis
fig, ax = matplot.subplots(figsize=(4, 4))
# Plot each column
for column in columns:
    ax.plot(x_data, dataFrame[column], label=column)
# Set title and legend
ax.set_title('Trending Graph')
ax.legend()
```

```python
# Histogram 
dataFrame.plot.hist(subplots=True, layout=(2,2), figsize=(5, 5), bins=20)
```
