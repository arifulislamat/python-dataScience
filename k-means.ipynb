```python
# Import libary
import matplotlib.pyplot as matplot
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans
```

```python
# Generate random data
genData = make_circles(n_samples=100, noise=0.5)
```

```python
kCluster = KMeans(n_clusters=5)
kCluster.fit(genData[0])
```

```python
# Create a figure and axis 
fig, (ax1, ax2) = matplot.subplots(1, 2, sharey=True, figsize = (16,6))
ax1.set_title('K-Means')
ax1.scatter(genData[0][:,0],genData[0][:,1], c=kCluster.labels_)
ax2.set_title('Original')
ax2.scatter(genData[0][:,0],genData[0][:,1],c=genData[1], cmap="rainbow")
```
